```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
import imageio
```

```python
img=cv.imread("test.png")
img=cv.cvtColor(img,cv.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
buf=[img]
#选定区域的图像
target=img[200:400,600:800]
#plt.imshow(target)
move=5 #每帧移动速度
#先指定移动的直线 y=x
for i in range(30):
    tmp=img.copy()
    for x in range(200,400):
        for y in range(600,800):
            if(x+i>400):
                tmpx=x+i-200
            else:
                tmpx=x+i
            if(y-i<600):
                tmpy=y-i+200
            else:
                tmpy=y-i
            tmp[x,y]=img[tmpx,tmpy]
    buf.append(tmp)
    
gif=imageio.mimsave('test.gif',buf,'GIF',duration=0.1)
```
